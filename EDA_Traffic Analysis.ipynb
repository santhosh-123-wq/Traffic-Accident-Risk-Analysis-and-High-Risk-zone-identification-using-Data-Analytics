# Exploratory Data Analysis

# Project : Traffic Accident Risk Analysis and High-Risk Zone Identification Using Data Analytics

```python
import numpy as np
import pandas as pd
import scipy.stats as stat
```

```python
data = pd.read_csv(r"C:\Users\santh\Downloads\Processed_Traffic_data.csv")
data.head(5)
```

```python
numeric_cols = data.select_dtypes(include = np.number).columns
numeric_cols
```

```python
categorical_cols = data.select_dtypes(include = 'object').columns
categorical_cols
```

# First Moment: Central Tendency

# Mean:

```python
for col in numeric_cols:
    print(col, ":", data[col].mean())
```

# Median:

```python
for col in numeric_cols:
    print(col, ":", data[col].median())
```

# Mode:

```python
for col in numeric_cols:
    print(col, ":", data[col].mode())
```

```python
for col in categorical_cols:
    print(col, ":", data[col].mode())
```

# Minimum:

```python
for col in numeric_cols:
    print(col, ":", min(data[col]))
```

# Maximum:

```python
for col in numeric_cols:
    print(col, ":", max(data[col]))
```

# Second Moment: Dispersion

# Variance

```python
for col in numeric_cols:
    print(col, ":", data[col].var())
```

# Standard deviation:

```python
for col in numeric_cols:
    print(col, ":", data[col].std())
```

# Range:

```python
for col in numeric_cols:
    Range = max(data[col]) - min(data[col])
    print(col, ":", Range)
```

```python
from scipy.stats import skew, kurtosis
```

# Third Moment: Skewness

```python
for col in numeric_cols:
    Skewness = data[col].skew()
    print(col, ":", Skewness)
```

# Fourth Moment: Kurtosis

```python
for col in numeric_cols:
    Kurtosis = data[col].kurt()
    print(col, ":", Kurtosis)
```

# Summary table for four moments:

```python
Summary_Table = pd.DataFrame ({
    'Mean' : data[numeric_cols].mean(),
    'Median' : data[numeric_cols].median(),
    'Standard_deviation' : data[numeric_cols].std(),
    'Variance' : data[numeric_cols].var(),
    'Range' : data[numeric_cols].max() - data[numeric_cols].min(),
    'Skewness' : data[numeric_cols].skew(),
    'Kurtosis' : data[numeric_cols].kurt()
})
```
